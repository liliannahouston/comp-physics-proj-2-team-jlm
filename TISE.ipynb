To Model the TISE under conditions where V(x) is infinite for x < 0, and V(x)=x for x >= 0

```python
import matplotlib.pyplot as plt
import numpy as np
import math
import scipy
```

Initial Parameters

```python
x0=0 #Initial position of integration
V0=0 #Initial Potential function
dx=1 #Simulation time step
xf=2000 #End of Calculation

#Initial Conditions
I0 = 0
dI0 = 1
E = 1000

#Constants
A = 1
m = 9.109*10**-31
hbar = (6.626*10**-34)/(2*math.pi)


stepcount=int(xf/dx)+1 #Gives the integer number of steps rounded up.
```

Arrays for Data

```python
x = np.linspace(x0,xf,stepcount)
P = np.zeros([stepcount]) #Array for Potential Curve
I = np.zeros([stepcount])
I1 = np.zeros([stepcount]) #Array for Psi
dI = np.zeros([stepcount]) #Array for dPsi
dI1 = np.zeros([stepcount])
k = np.zeros([stepcount]) #Array for k
ddI = np.zeros([stepcount]) #Array for d^2Psi
```

Functions to Simulate

```python
x[0] = x0
P[0] = V0
I1[0] = I0
dI1[0] = dI0
#This formula came from Phy 13 Class 16
k[0] = math.sqrt(2*m*(E-P[0])/(hbar**2))

for i in range(1, stepcount):
    x[i] = x[i-1] + dx
    P[i] = x[i]
    if E-P[i] >=0:
       #This is the classically allowed sinusoidal region
        k[i] = math.sqrt(2*m*(E-P[i])/(hbar**2))
        I1[i] = A*math.sin(k[i]*x[i])
    else:
        #This is the exponential classically forbidden region
        #This should be a separate k but I didnt make another array
        k[i] = math.sqrt(2*m*(P[i]-E)/(hbar**2))
        #This is the left moving wave because it has to converge to 0 at infinity
        I1[i] = (math.e)**(-k[i]*x[i])
```

Plot the Wavefunction

```python
plt.plot(x,I1)
plt.title('Wavefunction vs Position?')
plt.xlabel('Position (m)')
plt.ylabel('Wavefunction(x) (m^-1/2)')
plt.show()
```
