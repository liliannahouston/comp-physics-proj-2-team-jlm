```python
import matplotlib.pyplot as plt
import numpy as np
import math
import scipy
```

```python
x0=0
V0=0
dx=0.001
xf=10
guess = 1
I0=0
dI0=1
E=19

stepcount=int(xf/dx)+1
```

```python
x = np.linspace(x0,xf,stepcount)
P = np.zeros([stepcount]) #Array for Potential Curve
I = np.zeros([stepcount])
dI = np.zeros([stepcount]) #Array for dPsi
ddI = np.zeros([stepcount]) #Array for d^2Psi
```

```python
x[0]=x0
P[0]=V0
I[0]=I0
dI[0]=guess
ddI[0]=0

for i in range(1, stepcount):
    x[i] = x[i-1] + dx
    P[i] = x[i]
    ddI[i] = (2*P[i-1]-E)*I[i-1]
    dI[i] = ddI[i-1]*dx+dI[i-1]
    I[i] = dI[i-1]*dx+I[i-1]
```

```python
plt.plot(x,I)
plt.title('Wavefunction vs Position?')
plt.xlabel('Position (m)')
plt.ylabel('Wavefunction(x) (m^-1/2)')
plt.show()
```
